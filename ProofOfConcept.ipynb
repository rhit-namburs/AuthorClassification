```python
import os
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,GlobalAveragePooling1D,Embedding
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
fileloc = "./PierreTranTrain/"
articles_train = []
names_train = []
for filename in os.listdir(fileloc):
   with open(os.path.join(fileloc, filename), 'r') as f:
        articles_train.append(f.read())
        names_train.append(0)
```

```python
fileloc = "./RobinSidelTrain/"
for filename in os.listdir(fileloc):
   with open(os.path.join(fileloc, filename), 'r') as f:
        articles_train.append(f.read())
        names_train.append(1)
```

```python
fileloc = "./PierreTranTest/"
articles_test = []
names_test = []
for filename in os.listdir(fileloc):
   with open(os.path.join(fileloc, filename), 'r') as f:
        articles_test.append(f.read())
        names_test.append(0)
```

```python
fileloc = "./RobinSidelTest/"
for filename in os.listdir(fileloc):
   with open(os.path.join(fileloc, filename), 'r') as f:
        articles_test.append(f.read())
        names_test.append(1)
```

```python
tokenizer = Tokenizer(num_words = 100,oov_token="<OOV>")

# fitting the sentences to using created tokenizer object
tokenizer.fit_on_texts(articles_train)

# the full list of words is available as the tokenizer's word index
word_index = tokenizer.word_index
```

```python
train_sequences = tokenizer.texts_to_sequences(articles_train)
test_sequences = tokenizer.texts_to_sequences(articles_test)
```

```python
vocab_size = 10000
max_length = 100
trunc_type='post'
padding_type='post'
oov_tok = "<OOV>"
training_size = 20000


tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_tok)
# fitting tokenizer only to training set
tokenizer.fit_on_texts(articles_train)

word_index = tokenizer.word_index

# creating training sequences and padding them
traning_sequences = tokenizer.texts_to_sequences(articles_train)
train_padded = pad_sequences(traning_sequences,maxlen = max_length,
                                padding = padding_type,
                                truncating=trunc_type)

# creating  testing sequences and padding them using same tokenizer
testing_sequences = tokenizer.texts_to_sequences(articles_test)
test_padded = pad_sequences(testing_sequences,maxlen = max_length,
                                padding = padding_type,
                                truncating=trunc_type)

import numpy as np
# converting all variables to numpy arrays, to be able to work with tf version 2
train_padded = np.array(train_padded)
train_labels = np.array(names_train)
test_padded = np.array(test_padded)
test_labels = np.array(names_test)
```

```python
embedding_dim = 16

# creating a model for sentiment analysis
model  = Sequential([
                # addinging an Embedding layer for Neural Network to learn the vectors
                Embedding(vocab_size, embedding_dim, input_length = max_length),
                # Global Average pooling is similar to adding up vectors in this case
                GlobalAveragePooling1D(),
                Dense(24, activation = 'relu'),
                Dense(2,activation='softmax')
])

model.compile(loss = 'binary_crossentropy', optimizer = Adam(), metrics = ['accuracy'])


num_epochs = 100

history = model.fit(train_padded,train_labels, epochs = num_epochs,
                    validation_data = (test_padded,test_labels))
```
